### Spectrophotometric Macroscopic Species (Correct)

```python
# Create simulated macroscopic populations, assuming carboxylic acid protonation 
# won't create UV signal

# macro state 1 = microstate 1 + microstate 2
macro_pop1 = system2.populations[6] + system2.populations[7]
# macro state 2 = microstate 3 + microstate 5
macro_pop2 = system2.populations[5] + system2.populations[4]
# macro state 3 = microstate 4 + microstate 6
macro_pop3 = system2.populations[3] + system2.populations[2]
# macro state 4 = microstate 7 + microstate 8
macro_pop4 = system2.populations[1] + system2.populations[0]

system3_populations = np.stack((macro_pop1, macro_pop2, macro_pop3, macro_pop4), axis=0)
print(system3_populations.shape)

# Titration curves from populations

system3 = TitrationCurve.from_populations(system3_populations, pHrange)
plot3 = system3.plot("population") + system3.plot("free_energy")
plot3 = plot3.redim.range(population = (-0.1, 1.1))
plot3
```
